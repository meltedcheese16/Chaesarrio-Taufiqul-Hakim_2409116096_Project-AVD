3. Mengubah Data

```python
results = []

cols = df.select_dtypes(include=['float64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
columns_to_impute = ["1. What is your age?"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

melakukan import yang diperlukan dalam proses visualisasi data

### Menggunakan Scatterplot untuk mengetahui korelasi hubungan waktu penggunaan media sosial dan tingkat depresi

```python

plt.figure(figsize=(8, 4))
sns.scatterplot(x=df["8. What is the average time you spend on social media every day?"],
                y=df["18. How often do you feel depressed or down?"],
                alpha=0.1, color="blue")

plt.title("Hubungan Waktu Penggunaan Media Sosial dan Tingkat Depresi")
plt.xlabel("Waktu Penggunaan Media Sosial (jam)")
plt.ylabel("Tingkat Depresi")
plt.show()
correlation = df["8. What is the average time you spend on social media every day?"].corr(
                df["18. How often do you feel depressed or down?"])
print("Korelasi antara penggunaan media sosial dan depresi:", correlation)
```

### Menggunakan Scatterplot untuk mengetahui lebih lanjut waktu penggunaan media sosial dan tingkat depresi

# Insight

## Hasil Korelasi

Korelasi Awal: 0.0956 (Hampir Tidak Ada Hubungan)

Korelasi Lanjutan: 0.2401 (Hubungan Lemah-Sedang)

- Awalnya, hubungan antara penggunaan media sosial dan depresi sangat lemah (0.0956).

- Namun, setelah analisis lebih lanjut, hubungan tersebut menjadi lebih jelas (0.2401), meskipun tetap dalam kategori lemah hingga sedang.

- Ini menunjukkan bahwa penggunaan media sosial mungkin berhubungan dengan depresi, tetapi tidak menjadi satu-satunya faktor penyebab.

## Insight dari Hasil

### Hubungan positif, tetapi tidak kuat

- Semakin lama seseorang menggunakan media sosial, semakin tinggi kemungkinan mereka mengalami depresi.

- Namun, korelasi masih dalam kategori lemah hingga sedang, artinya media sosial bukan satu-satunya penyebab depresi.


### Kemungkinan Faktor Lain yang Berperan


- Konten yang dikonsumsi â†’ Jika seseorang lebih banyak melihat konten negatif atau membandingkan diri dengan orang lain, efeknya bisa lebih besar.
